#Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ["python", "c", "c++", "java"]
y = [85, 70, 60, 82]
plt.xlabel("Language")
plt.ylabel("No")
plt.title("wsCube", fontsize = 15)
c = ["y", "b", "m", "g"]
plt.bar(x, y, width = 0.2, color = c, align = "edge", edgecolor = 'r', linewidth = 4, linestyle = ":", alpha = 0.7)
plt.show()
```

Double bar graph


```python
x = ["python", "c", "c++", "java"]
y = [85, 70, 60, 82]
z = [65, 45, 34, 52]
plt.xlabel("Language")
plt.ylabel("No")
width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]
plt.title("wsCube", fontsize = 15)
plt.bar(p, y, width, color = 'r', align = "edge", alpha = 0.9)
plt.bar(p1, z, width, color = 'y', align = "edge", alpha = 0.8)
plt.xticks(p + width / 2, x, rotation = 10)
plt.show()
```

horaizontal bar graph

```python
x = ["python", "c", "c++", "java"]
y = [85, 70, 60, 82]
z = [65, 45, 34, 52]
plt.xlabel("Language")
plt.ylabel("No")
width = 0.2
p = np.arange(len(x))
p1 = [j + width for j in p]
plt.title("wsCube", fontsize = 15)
plt.barh(p, y, width, color = 'r', align = "edge", alpha = 0.9)
plt.barh(p1, z, width, color = 'y', align = "edge", alpha = 0.8)
plt.xticks(p + width / 2, x, rotation = 10)
plt.show()
```

#Scatter Plot

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 1, 4, 5, 3, 6]
plt.scatter(x, y, color = 'y', s = 100)
plt.title('Scatter Plot', fontsize = 15)
plt.xlabel('Day', fontsize = 10)
plt.ylabel('No.', fontsize = 10)
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 1, 4, 5, 3, 6]
size = [400, 100, 300, 200, 100, 400, 200]
c = ['r', 'y', 'b', 'g', 'r', 'y', 'g']
plt.scatter(x, y, color = c, s = size, marker = '*', edgecolor = 'y', linewidth = 1)
plt.title('Scatter Plot', fontsize = 15)
plt.xlabel('Day', fontsize = 10)
plt.ylabel('No.', fontsize = 10)
plt.colorbar()
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 1, 4, 5, 3, 6]
size = [400, 100, 300, 200, 100, 400, 200]
plt.scatter(x, y, cmap = 'viridis', s = size, marker = '*', edgecolor = 'y', linewidth = 1)
plt.title('Scatter Plot', fontsize = 15)
plt.xlabel('Day', fontsize = 10)
plt.ylabel('No.', fontsize = 10)
plt.colorbar()
plt.show()
```

#Histogram plot

---



```python
x = np.random.randint(10, 60, (50))
print(x)
```

```python
no = [59, 17, 30, 26, 36, 31, 33, 11, 43, 50, 30, 24, 15, 30, 29, 16, 36, 55, 38, 46, 45, 27, 47, 57,
 38, 43, 32, 10, 20, 23, 24, 33, 52, 22, 19, 38, 55, 40, 28, 23, 25, 39, 50, 50, 35, 47, 11, 12,
 13, 23]
```

```python
plt.hist(no, color = 'g', edgecolor = 'y', bottom = 10)
plt.title("go")
plt.xlabel('ML', fontsize = 10)
plt.ylabel('No.', fontsize = 10)
plt.show()
```

```python
plt.hist(no, color = 'g', edgecolor = 'y', histtype = "step")
plt.title("go")
plt.xlabel('ML', fontsize = 10)
plt.ylabel('No.', fontsize = 10)
plt.show()
```

```python
plt.hist(no, color = 'g', edgecolor = 'y', bottom = 10, orientation = 'horizontal')
plt.title("go")
plt.xlabel('ML', fontsize = 10)
plt.ylabel('No.', fontsize = 10)
plt.show()
```

```python
plt.hist(no, color = 'g', edgecolor = 'y', log = True)
plt.title("go")
plt.xlabel('ML', fontsize = 10)
plt.ylabel('No.', fontsize = 10)
plt.show()
```

#Pie Chart

```python
x = [10, 20, 30, 40]
y = ['C', 'C++', 'Python', 'Java']
plt.pie(x, labels = y)
plt.show()
```

```python
x = [10, 20, 30, 40]
y = ['C', 'C++', 'Python', 'Java']
ex = [0.3, 0.0, 0.0, 0.0]
plt.pie(x, labels = y, explode = ex, autopct = "%0.1f%%", shadow = True, labeldistance = 1.1, startangle = 20,
        textprops = {"fontsize" : 10}, wedgeprops = {"linewidth" : 3, "edgecolor" : 'g'})
plt.title('Lang')
plt.show()
```

#Stem Plot

```python
x = [1, 2, 3, 4, 5, 6] 
y = [2, 2, 5, 6, 4, 3]
plt.stem(x, y)
plt.show()
```

#Stack Plot

```python
x = [1, 2, 3, 4, 5]
area1 = [2, 3, 2, 5, 4]
area2 = [1, 2, 2, 6, 3]
area3 = [2, 3, 2, 4, 2]

l = ['area1', 'area2', 'area3']

plt.stackplot(x, area1, area2, area3, labels = l)
plt.legend(loc = 2)
plt.show()
```

#Step Plot

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 2, 3, 5]

plt.step(x, y, marker = "o", ms = 10)
plt.title('ML')
plt.grid()
plt.show()
```

#Fill Between plot

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.plot(x, y, color = 'r')
plt.fill_between(x = [2, 4], y1 = 2, y2 = 4)
```
