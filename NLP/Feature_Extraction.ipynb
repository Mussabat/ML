#Bag of Words

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'text' : ['people watch campusx', 'campusx watch campusx', 'people write comment', 'campusx write comment'], 'output' : [1, 1, 0, 0]})
```

```python
#vocab
print(cv.vocabulary_)
```

#Ngrams

```python
df = pd.DataFrame({'text' : ['people watch campusx', 'campusx watch campusx', 'people write comment', 'campusx write comment'], 'output' : [1, 1, 0, 0]})
```

```python
cv1 = CountVectorizer(ngram_range = (2, 2))
```

```python
ngrams = cv1.fit_transform(df['text'])
```

```python
print(cv1.vocabulary_)
```

```python
cv2 = CountVectorizer(ngram_range = (1, 2))
```

```python
ngrams1 = cv2.fit_transform(df['text'])
```

```python
print(cv2.vocabulary_)
```

```python
cv3 = CountVectorizer(ngram_range = (1, 3))
```

```python
ngrams2 = cv3.fit_transform(df['text'])
```

```python
print(cv3.vocabulary_)
```

```python
print(len(cv3.vocabulary_))
```

#Tf-Idf

```python
df = pd.DataFrame({'text' : ['people watch campusx', 'campusx watch campusx', 'people write comment', 'campusx write comment'], 'output' : [1, 1, 0, 0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
tfidf.fit_transform(df['text']).toarray()
```

```python
print(tfidf.idf_)
```

```python
print(tfidf.get_feature_names_out())
```
